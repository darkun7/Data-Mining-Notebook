```python
# 1.	Masukkan iris.csv ke dataframe
df = pd.read_csv('iris.csv')
# df.dropna(inplace = True)
df
```

```python
# 2.	Normalisasikan setiap attribut iris ke dataframe menggunakan normalisasi min max
def normalisasi(dataframe):
  variety = dataframe['variety']
  dataframe = dataframe.iloc[:, :-1]
  new_df = dataframe.copy()
  df_max = []
  df_min = []
  max = dataframe.max()
  min = dataframe.min()
  # max = dataframe.max().max() #global min-max
  # min = dataframe.min().min() #global min-max
  # new_df= (dataframe-min)/(max-min) #global min-max
  for _ in dataframe.columns:
    max = dataframe[_].max()
    min = dataframe[_].min()
    df_max.append(max)
    df_min.append(min)
    new_df[_] = (dataframe[_]-min)/(max-min)
  # return { "data" : new_df.join(variety),
  #          "max"  : max,
  #          "min"  : min, 
  #         }
  return { "data" : new_df.join(variety),
           "max"  : pd.Series(df_max, index = dataframe.columns),
           "min"  : pd.Series(df_min, index = dataframe.columns)
          }

df_norm = normalisasi(df)["data"]
df_norm
```

```python
# 3.	Cari centroid dari data yang telah dinormalisasi
def centroid(dataframe, variety):
  df_variety = dataframe[dataframe["variety"] == variety]
  df_variety = df_variety.iloc[:, :-1]
  centroid = df_variety.mean()
  return centroid

centroid_setosa     = centroid(df_norm,"Setosa")
centroid_versicolor = centroid(df_norm,"Versicolor")
centroid_virginica  = centroid(df_norm,"Virginica")

ptag.sep("Setosa")
print(centroid_setosa)
ptag.sep("Versicolor")
print(centroid_versicolor)
ptag.sep("Virginica")
print(centroid_virginica)
```

```python
# 4.	Masukkan iristesting.csv ke dataframe. Normalisasikan iristesting.csv menggunakan vmin dan vmax pada data 1
df_test = pd.read_csv('iristesting.csv')
# df_test.dropna(inplace = True)
df_test.head()
```

```python
min_train = normalisasi(df)["min"]
max_train = normalisasi(df)["max"]
def testing_normalisasi(dataframe):
  variety = dataframe['variety']
  dataframe = dataframe.iloc[:, :-1]
  new_df = dataframe.copy()
  # new_df = (dataframe-min_train)/(max_train-min_train) #global max
  for _ in dataframe.columns:
    max = max_train[_]
    min = min_train[_]
    new_df[_] = (dataframe[_]-min)/(max-min)
    new_df[_] = (dataframe[_]-min)/(max-min)
  return new_df.join(variety)

norm_test = testing_normalisasi(df_test)
norm_test
```

```python
# 5.	Prediksilah kelas data di iristesting.csv menggunakan kriteria kedekatan distance antara setiap data di iristesting.csv
#     ke centroid masing2 class di iris. Gunakan euclidean distance 
def distance(dataframe, centroid):
  d = 0
  for _ in centroid.index:
    d += (dataframe[_]-centroid[_])**2 
  return d**(1/2)

setosa_pred     = []
versicolor_pred = []
virginica_pred  = []
y = norm_test["variety"] #Y = actual result, X = predicted result
for i, row in norm_test.iterrows():
  setosa_temp     = distance(row,centroid_setosa)
  versicolor_temp = distance(row,centroid_versicolor)
  virginica_temp  = distance(row,centroid_virginica)
  
  setosa_pred.append(setosa_temp)
  versicolor_pred.append(versicolor_temp)
  virginica_pred.append(virginica_temp)
  
x = {"Setosa": setosa_pred,
     "Versicolor": versicolor_pred,
     "Virginica": virginica_pred}
x
```

```python
def predict(x,y):
  key = list(x.keys())
  record = len( x[key[0]] )
  result = []
  # print(key)
  tp = 0 #true positive
  # tn = 0 #true negative
  # fn = 0 #false negative
  fp = 0 #false positives
  for _ in range(record): #29
    compare = []
    for __ in x.keys():
      compare.append(x[__][_])
    idx = compare.index(min(compare))
    pred = key[idx]
    result.append(pred)

    if pred == y[_]:
      print(pred,"true")
      tp +=1
    else:
      print("false",pred,"expect", y[_])
      fp +=1
  return { "result": result,
            "correct": tp,
            "wrong": fp
          }
    #[0.046507516217964916, 0.7710845602805618, 1.1310130261815627]
    # break
pred_result = predict(x,y)
```

```python
# 6.	Hitunglah berapa persen klasifikasi yang benar (akurasi) saat memprediksi class 
correct = pred_result['correct']
wrong   = pred_result['wrong']
acuracy = correct/(correct+wrong)*100
print(correct, wrong, acuracy)
```
