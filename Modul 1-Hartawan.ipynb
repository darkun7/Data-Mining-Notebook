```
# SOAL D
# Tambahkan angka 1 pada setiap bilangan di baris ganjil array berikut. 
# kemudian tambahkan angka 2 pada setiap bilangan di kolom genap array berikut.
matrix_2d = np.array([
                        [1,8,8,2],
                        [8,2,1,2],
                        [10,1,4,7],
                        [1,4,7,8],
                        [8,2,3,4]])
sep("Orgiginal")
print(matrix_2d)
sep("Add 1 every Row")
matrix_2d[::2] +=1 #baris ganjil
print(matrix_2d)
sep("Add 2 every Column")
matrix_2d[:, 1::2]+=2 #genap
print(matrix_2d)

```

```
# SOAL E
# Baliklah sumbu z array 3D berikut
matrix_2e = np.array([
                      [[7,8,5], [4,7,7],[7,1,9],[5,4,10]],
                      [[2,1,2],[9,8,5],[3,7,3],[4,9,9]],
                      [[2,9,3],[2,4,8],[1,6,6],[8,1,3]]])
sep("Orgiginal")
print(matrix_2e)
sep("Transpose")
print(matrix_2e.T)
sep("Transpose alternative")
print(np.transpose(matrix_2e))
sep("z-transpose")
print(matrix_2e.transpose(2,0,1))
sep()
print(matrix_2e.transpose(2,1,0))
```

```
# KODE F
# Ubahlah array pada soal e menjadi array 2D berukuran 6 x 6 (hint : gunakan perintah reshape)
sep("Original")
print(matrix_2e)
sep("Sliced")
print(matrix_2e.reshape([6,6]))
```
